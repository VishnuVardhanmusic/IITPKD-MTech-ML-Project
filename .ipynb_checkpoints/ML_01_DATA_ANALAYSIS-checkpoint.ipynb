Deriving crop specific statistical parameters of the features

```python
import ipywidgets
from ipywidgets import interact

@interact
def summary(crops = list(df['label'].value_counts().index)):
    x = df[df['label'] == crops]
    print("---------------------------------------------")
    print("Statistics for Nitrogen")
    print("Minimum Nitrigen required : {0:.2f}".format(x['N'].min()))
    print("Average Nitrogen required : {0:.2f}".format(x['N'].mean()))
    print("Maximum Nitrogen required : {0:.2f}".format(x['N'].max())) 
    print("---------------------------------------------")
    print("Statistics for Phosphorous")
    print("Minimum Phosphorous required : {0:.2f}".format(x['P'].min()))
    print("Average Phosphorous required : {0:.2f}".format(x['P'].mean()))
    print("Maximum Phosphorous required : {0:.2f}".format(x['P'].max()))
    print("---------------------------------------------")
    print("Statistics for Potassium")
    print("Minimum Potassium required : {0:.2f}".format(x['K'].min()))
    print("Average Potassium required : {0:.2f}".format(x['K'].mean()))
    print("Maximum Potassium required : {0:.2f}".format(x['K'].max())) 
    print("---------------------------------------------")
    print("Statistics for Temperature")
    print("Minimum Temperature required : {0:.2f}".format(x['temperature'].min()))
    print("Average Temperature required : {0:.2f}".format(x['temperature'].mean()))
    print("Maximum Temperature required : {0:.2f}".format(x['temperature'].max()))
    print("---------------------------------------------")
    print("Statistics for Humidity")
    print("Minimum Humidity required : {0:.2f}".format(x['humidity'].min()))
    print("Average Humidity required : {0:.2f}".format(x['humidity'].mean()))
    print("Maximum Humidity required : {0:.2f}".format(x['humidity'].max()))
    print("---------------------------------------------")
    print("Statistics for PH")
    print("Minimum PH required : {0:.2f}".format(x['ph'].min()))
    print("Average PH required : {0:.2f}".format(x['ph'].mean()))
    print("Maximum PH required : {0:.2f}".format(x['ph'].max()))
    print("---------------------------------------------")
    print("Statistics for Rainfall")
    print("Minimum Rainfall required : {0:.2f}".format(x['rainfall'].min()))
    print("Average Rainfall required : {0:.2f}".format(x['rainfall'].mean()))
    print("Maximum Rainfall required : {0:.2f}".format(x['rainfall'].max()))
```

CROP SUGGESTION BASED ON INPUT SOIL/CLIMATE PARAMETER

```python
@interact
def compare(conditions = ['N','P','K','temperature','ph','humidity','rainfall']):
    print("Average Value for", conditions,"is {0:.2f}".format(df[conditions].mean()))
    print("Crops which require MORE amount of ", conditions,' (greater than avg)\n')
    print(df[df[conditions] > df[conditions].mean()]['label'].unique())
    print((df[df['label'] == 'chickpea'][conditions].mean()))
    print("----------------------------------------------")
    print("Crops which require LESS amount of ", conditions,' (lower than avg)\n')
    print(df[df[conditions] != df[conditions].mean()]['label'].unique())
    print((df[df['label'] == 'chickpea'][conditions].mean()))
```

DERIVING THE NPK RATIOS FOR THE PROMINENT CROPS OF INDIA

```python
crop_summ = pd.pivot_table(df,index=['label'],aggfunc='mean')
crop_summ.head()
```

```python
fig = plt.figure(figsize=(25,10))
plt.subplot(2,4,1)
rice_npk = crop_summ[crop_summ.index=='rice']
values = [rice_npk['N'][0], rice_npk['P'][0], rice_npk['K'][0]]
plt.pie(values,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("RICE")

plt.subplot(2,4,2)
kid_npk = crop_summ[crop_summ.index=='kidneybeans']
kvalues = [kid_npk['N'][0], kid_npk['P'][0], kid_npk['K'][0]]
plt.pie(kvalues,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("KIDNEYBEANS")

plt.subplot(2,4,3)
pe_npk = crop_summ[crop_summ.index=='pigeonpeas']
pvalues = [pe_npk['N'][0], pe_npk['P'][0], pe_npk['K'][0]]
plt.pie(pvalues,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("PIGEONPEAS")

plt.subplot(2,4,4)
ap_npk = crop_summ[crop_summ.index=='apple']
avalues = [ap_npk['N'][0], ap_npk['P'][0], ap_npk['K'][0]]
plt.pie(avalues,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("APPLE")

plt.subplot(2,4,5)
pa_npk = crop_summ[crop_summ.index=='papaya']
pavalues = [pa_npk['N'][0], pa_npk['P'][0], pa_npk['K'][0]]
plt.pie(pavalues,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("PAPAYA")

plt.subplot(2,4,6)
coc_npk = crop_summ[crop_summ.index=='coconut']
ccvalues = [coc_npk['N'][0], coc_npk['P'][0], coc_npk['K'][0]]
plt.pie(ccvalues,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("COCONUT")

plt.subplot(2,4,7)
jut_npk = crop_summ[crop_summ.index=='jute']
jvalues = [jut_npk['N'][0], jut_npk['P'][0], jut_npk['K'][0]]
plt.pie(jvalues,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("JUTE")

plt.subplot(2,4,8)
cof_npk = crop_summ[crop_summ.index=='coffee']
cvalues = [cof_npk['N'][0], cof_npk['P'][0], cof_npk['K'][0]]
plt.pie(cvalues,labels=['Nitrogen (N)','Phosphorous (P)','Potassium (K)'],autopct = '%1.1f%%')
plt.title("COFFEE")

plt.show()
```
